```python
import pandas as pd

#load dataset
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\Data Analysis And Automation\Sample Data\data_jobs.csv")
df.head()

# cleaning the dataset
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df
```

```python
df_kenya = df[df['job_country'] == 'Kenya'].copy()
```

```python
df_kenya
```

```python
# Extract month and year from job_posted_date in verbage format

df_kenya['job_posted_month'] = df_kenya['job_posted_date'].dt.strftime('%B') #refer from datetime documentation for % sighns
```

```python
df_kenya
```

```python
# putting months along the index in correct order and job titles along the columns

df_kenya_pivot = df_kenya.pivot_table(index='job_posted_month', columns='job_title', aggfunc='count')
```

```python
df_kenya_pivot.reset_index(inplace=True)
df_kenya_pivot['month_no'] = pd.to_datetime(df_kenya_pivot['job_posted_month'], format='%B').dt.month
df_kenya_pivot.sort_values('month_no', inplace=True)
df_kenya_pivot.set_index('job_posted_month', inplace=True)
df_kenya_pivot.drop(columns='month_no', inplace=True)

df_kenya_pivot
```

```python
top_3 = df_kenya['job_title_short'].value_counts().head(3)
top_3.index.tolist()
```

```python
top_3 = top_3.index.tolist()
```

```python
import matplotlib.pyplot as plt


# aggregate counts across the top-level of the MultiIndex columns so job titles become columns
df_counts = df_kenya_pivot.groupby(level=1, axis=1).sum()

# plot the top 3 job titles
df_counts[top_3].plot(kind='line', marker='o')
plt.title('Monthly job postings for top 3 job titles in Kenya')
plt.ylabel('Count')
plt.xlabel('2023 Months')
plt.show()
```
