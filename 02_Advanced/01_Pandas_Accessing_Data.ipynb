```python

import pandas as pd

#load dataset
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\Data Analysis And Automation\Sample Data\data_jobs.csv")
df.head()

# cleaning the dataset
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.iloc[0:10, 5]  # accessing rows by integer location 0 to 9 and column 5 _ index based
```

```python
# Multiple row and column selection with .loc
# first part defines the rows we want to select
# second part defines the columns we want to select
# accessing rows by label 0 to 10 _ index based
# note that in loc the last index is included unlike iloc 
df.loc[ : , ['salary_rate', 'salary_year_avg', 'salary_hour_avg']].dropna(subset=['salary_rate'])  # dropna to remove rows with missing values, no indexing to get all rows
```
