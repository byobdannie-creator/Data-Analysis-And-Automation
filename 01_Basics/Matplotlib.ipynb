```python
# introduction to Matplotlib

# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization and provides a variety of plotting functions to create different types of charts and graphs.
# It is built on NumPy arrays and designed to work with the broader SciPy stack. Matplotlib offers a high degree of customization, allowing users to modify almost every aspect of a figure, including colors, labels, line styles, and more.
# Matplotlib is particularly useful for creating publication-quality figures and is often used in conjunction with other libraries such as Pandas and Seaborn for data analysis and visualization tasks.
# Some key features of Matplotlib include:
# 1. Versatile Plotting: Matplotlib supports a wide range of plot types, including line plots, scatter plots, bar charts, histograms, pie charts, and more.
# 2. Customization: Users can customize plots extensively, including axes, labels, titles, legends, and colors.
# 3. Integration: Matplotlib integrates well with other libraries like NumPy, Pandas, and Jupyter Notebooks, making it a popular choice for data analysis and visualization in Python.
# 4. Interactive Plots: Matplotlib can create interactive plots that allow users to zoom, pan, and update visualizations dynamically.
# 5. Export Options: Plots created with Matplotlib can be saved in various formats, including PNG, PDF, SVG, and more.  
# To get started with Matplotlib, you typically import the library and use its functions to create and customize plots. Here's a simple example of creating a basic line plot:
```

```python
job_counts = df.job_title_short.value_counts()
job_counts.plot.barh()
```

```python
# get in the order that you want
```

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts.values)
```

```python
# cleaning up plots visualiation
# matplotlib provides various options to customize and clean up plots for better visualization. Here are some common techniques to enhance the appearance of your plots:    
# 1. Titles and Labels: Add titles and axis labels to provide context to your plots.
# 2. Gridlines: Enable gridlines to improve readability.
# 3. Legends: Include legends to differentiate between multiple data series.
```

```python
plt.bar(job_counts.index, job_counts)
plt.title("Posting by Job Title")
plt.show()
```

```python
# add a ylabel to indicate what the values represent

plt.bar(job_counts.index, job_counts)
plt.title("Posting by Job Title")
plt.ylabel("Number of Job Postings")
#clean up x axis labels if necessary
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job_counts.plot(kind='bar')
plt.title("Posting by Job Title")
plt.ylabel("Number of Job Postings")
plt.xlabel("")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Exercise: Visualise Salaries
# Visualizing salaries can provide insights into the distribution and trends of salaries in a dataset. Here are some common ways to visualize salary data using Matplotlib:

#for this exercise, we are going to find the median salary based on job title itself to determine high paying jobs and low paying jobs.
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values # automatically sorts the values in ascending order
```

```python
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values
```

```python
#plotting the salary based on job title
# assign the median salaries to a variable
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values().dropna()
```

```python
job_salary.plot(kind='barh')
plt.xlabel('Salary($USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')
plt.show()
```
