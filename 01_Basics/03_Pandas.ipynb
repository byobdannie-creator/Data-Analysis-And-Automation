```python
# Exercise
# import dataset
import pandas as pd

#load dataset
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\Data Analysis And Automation\Sample Data\data_jobs.csv")
df.head()

# cleaning the dataset
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df['job_country'].value_counts().head(20)
```

```python
df['job_country'].isin(['Kenya']).any()
```

```python
# creating a dataframe filter for Kenya job listings only
kenya_jobs = df[df['job_country'] == 'Kenya']
```

```python
kenya_jobs
```

```python
#cleaning the kenya_jobs dataframe
kenya_jobs = kenya_jobs[kenya_jobs['salary_year_avg'].notna()]
```

```python
kenya_jobs
```

```python
kenya_jobs.groupby('job_title_short')[['salary_year_avg']].agg(['mean', 'median', 'min', 'max']).sort_values(by=('salary_year_avg', 'mean'), ascending=False)
```
